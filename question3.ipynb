### Image processing: read image and find the 5 first choices of the student

```python
import cv2
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
import numpy as np

path = './data/' + fn
image = imread(path)

imshow(image)
```

```python
crop = image[690:2800, 430:2100]
first5 = crop[0:300, 0:730]
# imshow(test[0:300, 870:1600])
first5 = cv2.cvtColor(first5, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(first5, 0, 255, cv2.THRESH_BINARY_INV)
imshow(thresh)
cnts, hierarychy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

choices = []

for cnt in cnts:
    area = cv2.contourArea(cnt)
    if area > 20:
        center = np.round(np.mean(cnt, axis=0)[0]).astype(int)
        choices.append(center)

choices.sort(key=lambda x:x[1])

```

### Display 5 first choices of the student

```python
for choice in choices:
    x = choice[0]
    if 200 < x < 300:
        print('A', end =' ')
    if 300 < x < 400:
        print('B', end =' ')
    if 400 < x < 500:
        print('C', end =' ')
    if 500 < x < 600:
        print('D', end =' ')
    if 600 < x < 700:
        print('E', end =' ')
```
